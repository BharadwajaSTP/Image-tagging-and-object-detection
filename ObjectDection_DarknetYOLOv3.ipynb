## **Step-05: Prepare Necessary Files for Training.**

```python
"""
# Objective:
#  Creating files train.txt and valid.txt for training in Darknet framework.
#
# Algorithm:
#  --> Setting up full paths.
#  --> List of paths.
#  --> Extracting 15% of paths to save into valid.txt file.
#  --> Remaining 85% of paths to save into train.txt file.
#  --> Writing paths into train and validation files.
#
# Result:
#  Files train.txt and valid.txt with full paths to images.
"""


# Importing needed libraries.
import os


# Global variables.
# Full or absolute path to the folder containing annotated image files and corresponding data files.
full_path_to_images = "/content/darknet/bdd100k"

# Define a list vaariable to write paths.
PathList = []


"""
Start of: Getting list of full paths to labelled images.
"""

# Check point.
# Getting the current directory.
print(os.getcwd())

# Changing the current directory to the one with images.
os.chdir(full_path_to_images)

# Check point.
# Getting the current directory.
print(os.getcwd())

# Using os.walk for going through all directories and files in them from the current directory.
# Fullstop in os.walk('.') means the current directory.
for current_dir, dirs, files in os.walk('.'):
    # Going through all files
    for file in files:
        # Checking if filename ends with '.jpg'
        if file.endswith('.jpg'):
            # Preparing path to save into train.txt file.
            path_to_save_into_txt_files = full_path_to_images + '/' + file

            # Appending the line into the list.
            # We use here '\n' to move to the next line when writing lines into txt files.
            PathList.append(path_to_save_into_txt_files + '\n')


# Slicing first 15% of elements from the list to write into the valid.txt file.
p_test = PathList[:int(len(PathList) * 0.15)]

# Deleting from initial list first 15% of elements.
PathList = PathList[int(len(PathList) * 0.15):]

"""
End of: Getting list of full paths to labelled images
"""


"""
Start of: Creating train.txt and valid.txt files.
"""

# Creating file train.txt and writing 85% of lines in it.
with open('train.txt', 'w') as train_txt:
    # Going through all elements of the list.
    for Element in PathList:
        # Writing current path at the end of the file.
        train_txt.write(Element)

# Creating file valid.txt and writing 15% of lines in it.
with open('valid.txt', 'w') as test_txt:
    # Going through all elements of the list.
    for Element in p_test:
        # Writing current path at the end of the file.
        test_txt.write(Element)

"""
End of: Creating train.txt and valid.txt files.
"""

print("Created train.txt and valid.txt files.")
```

```python
# Move to the 'root' directory of the 'darknet' framework.
%cd /content/darknet
```

```python
"""
# Objective:
#  Creating files labelled_data.data and classes.names for training in Darknet framework.
#
# Algorithm:
#  --> Setting up full paths.
#  --> Reading file classes.txt
#  --> Creating file classes.names
#  --> Creating file bdd100k.data
#
# Result:
#  Files classes.names and bdd100k.data needed to train in Darknet framework.
"""


# Importing needed libraries.
import os


# Global variables.
# Full or absolute path to the folder containing annotated image files and corresponding data files.
full_path_to_images = "/content/darknet/bdd100k"


"""
Start of: Creating file classes.names
"""

# Defining counter for classes.
Counter = 0

# Creating file classes.names from existing one classes.txt
with open(full_path_to_images + '/' + 'classes.names', 'w') as names, \
     open(full_path_to_images + '/' + 'classes.txt', 'r') as txt:
    # Going through all lines in txt file and writing them into names file.
    for line in txt:
        # Copying all info from file txt to names.
        names.write(line)

        # Increasing counter.
        Counter += 1

"""
End of: Creating file classes.names
"""


"""
Start of: Creating file labelled_data.data
"""

# Creating file labelled_data.data.
with open(full_path_to_images + '/' + 'bdd100k.data', 'w') as data:
    # Writing needed 5 lines i.e. Number of classes.
    # By using '\n' we move to the next line.
    data.write('classes = ' + str(Counter) + '\n')

    # Location of the train.txt file.
    data.write('train = ' + full_path_to_images + '/' + 'train.txt' + '\n')

    # Location of the valid.txt file.
    data.write('valid = ' + full_path_to_images + '/' + 'valid.txt' + '\n')

    # Location of the classes.names file.
    data.write('names = ' + full_path_to_images + '/' + 'classes.names' + '\n')

    # Location where to save weights.
    data.write('backup = backup')

"""
End of:
Creating file labelled_data.data
"""


print("Created classes.names and labelled_data.data files.")
```
